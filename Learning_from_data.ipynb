```python
import pandas as pd
cars = pd.read_csv("C:\\Users\\sg965\\Downloads\\sklearn\\cars_dataset.csv")
```

```python
cars.head()
```

```python
ax1 = cars.plot.scatter(x='weight', y='horsepower',grid=True)
ax1.set_title('Linear Regression Relation between Weight and Horsepower')
ax1.set_xlim(1500,5000)
ax1.set_ylim(40,240)
ax1.legend(title='data', loc='upper left')
```

```python
x = cars[['weight']]
y_data = cars['horsepower']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x,y_data)#
slope = reg.coef_
intercept = reg.intercept_
print(f'Slope of the Regression is {slope}')
print(f'Intercept of the Regression is {intercept}')
```

```python
y = reg.predict(x)
y_pred = y
cars['predicted_horsepower'] = y_pred
ax = cars.plot.scatter(x='weight', y='horsepower')
ax = cars.plot.line(x='weight', y='predicted_horsepower', ax=ax, c='red')
```

```python
R_sqr = reg.score(x,y)
print(f'R square value is {R_sqr}')
```
